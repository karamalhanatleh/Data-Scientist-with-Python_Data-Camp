```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../dataset/wine.csv')
df.head()
```

```python
sns.displot(df['Alcohol'], kde=True, bins=10)
```

```python
sns.displot(df['Alcohol'], kde=False, bins=10)
```

```python
sns.displot(df['Alcohol'], kde=True, bins=10)
```

```python
sns.displot(df['Alcohol'], kind='kde', rug=False, fill=False)
```

```python
df['alcohol'] = df['Alcohol']
```

```python
sns.displot(df['Alcohol'], kind='kde', rug=True, fill=True)
```

```python
sns.displot(df['alcohol'], kind='ecdf')
```
