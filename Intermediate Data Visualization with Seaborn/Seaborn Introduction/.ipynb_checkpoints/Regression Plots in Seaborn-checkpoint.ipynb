```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../dataset/wind_o.csv')
#df['alcohol'] = df['Alcohol']
df.head()
```

```python
sns.regplot(data=df, x="alcohol" , y="pH" )
```

### regplot - low level

```python
sns.regplot(data=df,
x="alcohol"
,
y="quality")
```

### lmplot - high leve

```python
sns.lmplot(data=df,
          x="alcohol" , y='quality')
```

```python
sns.lmplot(data=df,x="quality",y="alcohol",hue="type")
```

## Create a regression plot


```python
# Create a regression plot of premiums vs. insurance_losses
sns.regplot(x="insurance_losses",y = "premiums", data = df)




# Display the plot
plt.show()
```

```python
# Create an lmplot of premiums vs. insurance_losses
sns.lmplot(x="insurance_losses",y="premiums",data=df)

# Display the second plot
plt.show()
```

## Plotting multiple variables


```python
# Create a regression plot using hue
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           hue="Region")

# Show the results
plt.show()
```

## Facetting multiple regressions


```python
# Create a regression plot with multiple rows
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           row="Region")

# Show the plot
plt.show()
```
