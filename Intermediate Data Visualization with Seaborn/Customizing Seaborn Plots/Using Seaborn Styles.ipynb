# Using Seaborn Styles


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../dataset/2018 College Scorecard Tuition.csv')
#df['alcohol'] = df['Alcohol']
df['Tuition']
```

```python
sns.set()
df['Tuition'].plot.hist()
```

```python
for style in ['white','dark','whitegrid','darkgrid','ticks']:
    sns.set_style(style)
    sns.displot(df['Tuition'])
    plt.show()
```

```python
sns.set_style('white')
sns.displot(df['Tuition'])
sns.despine(left=False)
```

```python
sns.set_style('white')
sns.displot(df['Tuition'])
sns.despine(left=True)
```

```python
sns.set_style('white')
sns.displot(df['Tuition'])
sns.despine(left=False, right = False ,top=False)
```

## Setting the default style


```python
# Plot the pandas histogram
df['fmr_2'].plot.hist()
plt.show()
plt.clf()

# Set the default seaborn style
sns.set()

# Plot the pandas histogram again
df['fmr_2'].plot.hist()
plt.show()
plt.clf()
```

## Comparing styles


```python
# Plot with a dark style 
sns.set_style('dark')
sns.displot(df['fmr_2'])
plt.show()

# Clear the figure
plt.clf()
```

```python
plt.clf()
sns.set_style('whitegrid')
sns.displot(df['fmr_2'])
plt.show()

# Clear the figure
plt.clf()
```

## Removing spines


```python
# Set the style to white
sns.set_style('white')

# Create a regression plot
sns.lmplot(data = df,
           x = 'pop2010',
           y = 'fmr_2')

# Remove the spines
sns.despine()

# Show the plot and clear the figure
plt.show()
plt.clf()
```
