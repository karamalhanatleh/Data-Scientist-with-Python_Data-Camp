```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("datasets/rgb_array.npy","rb") as f:
    rgb_array = np.load(f)

plt.imshow(rgb_array)
plt.show()
```

```python
flipped_logo = np.flip(rgb_array)
plt.imshow(flipped_logo)

plt.show()
```

```python
flipped_logo = np.flip(rgb_array , axis = 0 )
plt.imshow(flipped_logo)

plt.show()
```

```python
flipped_logo = np.flip(rgb_array , axis = 1)
plt.imshow(flipped_logo)

plt.show()
```
